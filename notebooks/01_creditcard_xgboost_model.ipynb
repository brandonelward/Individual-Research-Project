```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loads the New Dataset
try:
    df_new = pd.read_csv('../data/creditcard.csv')
    print("New dataset loaded successfully!")
except FileNotFoundError:
    print("Error: Make sure the new dataset file is in the 'data' directory and the filename is correct.")

# Initial Exploration
print("\nFirst 5 rows of the new dataset:")
display(df_new.head())

print("\n Information:")
df_new.info()

print("\n Descriptive Statistics:")
display(df_new.describe())

print("\n Missing Value Count:")
print(df_new.isnull().sum())
```

```python
# Planning & Investigation

# Target column ('Class') - How many fraud cases there are
print("Distribution of the 'Class' column:")
print(df_new['Class'].value_counts())
```

```python
# Preprocessing Function

from sklearn.preprocessing import StandardScaler

def preprocess_credit_card_data(df):
    """
    Prepares the credit card dataset for model training.
    - Scales the 'Amount' column.
    - Drops the 'Time' column.
    - Separates features (X) from the target (y).
    """
    df_processed = df.copy()
    
    # Scale the 'Amount' feature, as its range is very different from the 'V' features
    scaler = StandardScaler()
    df_processed['scaled_Amount'] = scaler.fit_transform(df_processed['Amount'].values.reshape(-1, 1))
    
    # Drop the original 'Amount' and the 'Time' columns
    df_processed = df_processed.drop(['Time', 'Amount'], axis=1)
    
    # Separate the features (X) from the target (y)
    X = df_processed.drop('Class', axis=1)
    y = df_processed['Class']
    
    return X, y
```

```python
# Executes Preprocessing

X_new, y_new = preprocess_credit_card_data(df_new)

print("Preprocessing complete.")
print("\nShape of our new features (X_new):", X_new.shape)
print("Shape of our new target (y_new):", y_new.shape)

print("\nFirst 5 rows of the processed features (X_new):")
display(X_new.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from xgboost import XGBClassifier

# Uses 'stratify=y_new' to ensure the proportion of fraud cases is the same in both sets.
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(
    X_new, y_new, test_size=0.2, random_state=42, stratify=y_new
)

print("Data split complete.")
print("Shape of X_train_new:", X_train_new.shape)
print("Shape of X_test_new:", X_test_new.shape)


# Re-calculates the scale_pos_weight for the dataset
# This is crucial as the class imbalance is different.
ratio_new = (y_new == 0).sum() / (y_new == 1).sum()
print(f"\nNew class imbalance ratio (scale_pos_weight): {ratio_new:.2f}")


# Trains XGBoost model
print("\nTraining the model on the credit card dataset...")
xgboost_creditcard_model = XGBClassifier(
    objective='binary:logistic',
    scale_pos_weight=ratio_new,
    use_label_encoder=False,
    eval_metric='logloss'
)

xgboost_creditcard_model.fit(X_train_new, y_train_new)
print("Model training complete.")


# Evaluates the model's performance 
y_pred_new = xgboost_creditcard_model.predict(X_test_new)

print("\n Model Evaluation:")
print("\nClassification Report:")
# '1' represents Fraud, '0' represents Not Fraud
print(classification_report(y_test_new, y_pred_new, target_names=['Not Fraud (0)', 'Fraud (1)']))

print("\nConfusion Matrix:")
cm_new = confusion_matrix(y_test_new, y_pred_new)

# Visualise the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_new, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for New Dataset')
plt.show()
```

```python
import joblib
import shap

# Saves the New Model and its SHAP Explainer

# Saves the trained model
xgboost_creditcard_model_path = '../models/xgboost_creditcard_model.joblib'
joblib.dump(xgboost_creditcard_model, xgboost_creditcard_model_path)

print(f"Model saved successfully to: {xgboost_creditcard_model_path}")

# Creates and saves a SHAP explainer for the model
# Uses the training data (X_train_new) to create the explainer for consistency
print("\nCreating SHAP explainer...")
shap_explainer_xgboost_creditcard = shap.TreeExplainer(xgboost_creditcard_model)
print("Explainer created.")

# Saves the explainer
xgboost_creditcard_explainer_path = '../models/shap_explainer_xgboost_creditcard.joblib'
joblib.dump(shap_explainer_xgboost_creditcard, xgboost_creditcard_explainer_path)
print(f"New SHAP explainer saved successfully to: {xgboost_creditcard_explainer_path}")
```
